```python
import pandas as pd

names = pd.read_csv('names.csv')
```

```python
scotland = names[names['country'] == 'Scotland']
england = names[names['country'] == 'England']
wales = names[names['country'] == 'Wales']
ireland = names[names['country'] == 'Ireland']
```

```python
from pandasql import sqldf

pysqldf = lambda q: sqldf(q, globals())

complete_data = pysqldf("""SELECT england.year, scotland.boys AS scotland_boys, scotland.girls AS scotland_girls,
                  wales.boys AS wales_boys, wales.girls AS wales_girls, england.boys AS england_boys,
                  england.girls AS england_girls,ireland.boys AS ireland_boys, ireland.girls AS ireland_girls
                  FROM scotland 
                  INNER JOIN england ON scotland.year == england.year
                  INNER JOIN wales ON scotland.year == wales.year
                  INNER JOIN ireland ON scotland.year == ireland.year""")

missing_data = pysqldf("""SELECT ireland.year, scotland.boys AS scotland_boys, scotland.girls AS scotland_girls,
                  wales.boys AS wales_boys, wales.girls AS wales_girls, england.boys AS england_boys,
                  england.girls AS england_girls,ireland.boys AS ireland_boys, ireland.girls AS ireland_girls
                  FROM ireland 
                  LEFT JOIN england ON ireland.year == england.year
                  LEFT JOIN wales ON ireland.year == wales.year
                  LEFT JOIN scotland ON scotland.year == ireland.year""")
```
